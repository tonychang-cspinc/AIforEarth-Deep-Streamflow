```python
import folium 
import pandas as pd
import numpy as np
```

```python
gage_df = pd.read_csv('/datadrive/stream_data/waterdata.txt', header=40, sep='\t')
```

```python
gage_df.drop([0])
```

```python
locations = gage_df[['site_no','station_nm','dec_lat_va','dec_long_va']].drop([0]).reset_index()
```

```python
reference_locations = pd.DataFrame({'camera_nm':['BROWNS BROOK', 'PARKERS BROOK', 'PEAVINE CREEK'],\
                       'lat':[42.034099,42.394455, 41.948643],\
                       'lon':[-71.159741,-72.047279, -86.187894]})
```

```python
locationlist = locations[['dec_lat_va','dec_long_va']].values.tolist()
center_lat = np.mean(locations.dec_lat_va.astype('float'))
center_lon = np.mean(locations.dec_long_va.astype('float'))
```

```python
map = folium.Map(location=[center_lat, center_lon], zoom_start=10)
for point in range(len(locationlist)):
    folium.Marker(locationlist[point], popup=locations['station_nm'][point]).add_to(map)
for pnt in range(len(reference_locations)):
    folium.Marker((reference_locations['lat'][pnt],reference_locations['lon'][pnt]),\
                 popup=reference_locations['camera_nm'][pnt], \
                 icon=folium.Icon(color='red', icon_color='white')).add_to(map)
map
```
